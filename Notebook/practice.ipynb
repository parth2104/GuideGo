```python
from langchain_openai import ChatOpenAI
import os

llm=ChatOpenAI(
    model="gpt-5-mini-2025-08-07",
    api_key=os.getenv("api_key"),
    openai_api_base=os.getenv("endpoint")
)
```

```python
llm.invoke("hello brother what is  your name and how old are you?")
```
